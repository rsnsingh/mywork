```python
# 6.15 reverse binary
#=========================
rev=""     # create an empty string
n = int(input())  
binary = bin(n)[2:] # generate binary number for an inputed number n
i = len(binary) - 1 # get the last index value, i.e length of string - 1
while(i >=0 ):      # loop till index i is greater than or equal to 0, start the loop from last element index
    rev = rev + binary[i]    # add each element from last one by one to new string rev 
    i = i - 1       # decrement the index i by one 
print(rev)          # print reversed binary number
```

```python
# 5.18  Season printing
#======================================
month = input('Enter month name : ')
day   = int(input('Enter Day of Month : '))
months = ['jan','feb','march','april','may','june','july','august','sep','oct','nov','dec']

Spring = ['march','april','may','june']
Summer = ['june','july','august','sep']
Autumn = ['sep','oct','nov','dec']
Winter = ['dec','jan','feb','march']

if month in months:   
    if day > 0 and day <=31:
        if month in Spring:
            if (Spring.index(month)==0 and day >=20 ) or (Spring.index(month)==3 and day <=20):  
                season ="Spring"
            else:
                season = "Spring"
        else:        
            if month in Summer:
                if (Summer.index(month)==0 and day >=21 ) or (Summer.index(month)==3 and day <=21):
                    season ="Summer"
                else:
                    season = "Summer"
            else:
                if month in Autumn:
                    if (Autumn.index(month)==0 and day >=22 ) or (Autumn.index(month)==3 and day <=20):
                        season ="Autumn"
                    else:
                        season ="Autumn"
                else:
                    if month in Winter:
                        if (Winter.index(month)==0 and day >=21 ) or (Winter.index(month)==3 and day <=19):
                            season ="Winter"
                        else:
                            season ="Winter"
    else:
        season = "Invalid"
else:
    season = "Invalid"
print(season)
```

```python
# 5.15  Remove Gray color
#================================
color_rgb =[]
color_new =[]
print(" Input r g b  values ")
color_rgb.append(int(input())) # red value
color_rgb.append(int(input())) # green value
color_rgb.append(int(input())) # blue value
print("Original Color :",color_rgb)

gray = min(color_rgb) # minimum from list, gray color factor

print("Gray value in color :",gray)  # Generate New color list without gray 
color_new.append(color_rgb[0] - gray) # subtract gray from red color
color_new.append(color_rgb[1] - gray) # subtract gray from green color
color_new.append(color_rgb[2] - gray) # subtract gray from blue color

print("Color without Gray : ", color_new)
```

```python
# 5.19   Exact Change 
#========================
val = int(input())
d=0
q=0
di=0
n=0
p=0
while(val>0):
    if(val>=100):
        d=d+1
        val=val-100
    else:
        if(val>25):
            q=q+1
            val=val-25
        else:
            if(val>=10):
                di=di+1
                val=val-10
            else:
                if(val>5):
                    n=n+1
                    val=val-5
                else:
                    if(val>1):
                        p=p+1
                        val=val-1    

if(d>0):
    if(d==1):
        print(d,"Dollar")
    else:    
        print(d,"Dollars")

if(q>0):
    if(q==1):
        print(q,"Quarter")
    else:    
        print(q,"Quarters")

if(di>0):
    if(di==1):
        print(di,"Dime")
    else:    
        print(di,"Dimes")

if(n>0):
    if(n==1):
        print(n, "Nickel")
    else:    
        print(n,"Nickels")
if(p>0):
    if(p==1):
        print(p, "Pennie")
    else:    
        print(p,"Pennies")
```
